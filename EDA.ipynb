```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Abrindo a base de dados .csv
df_base= pd.read_csv('base_limpa.csv')

#Realizando uma rápida observação dos dados
df_base.head(10)
```

```python
df_base.describe()
```

```python
sns.lineplot(data=df_base, x="ano_final", y="PREÇO MÉDIO REVENDA")
```

```python
sns.lineplot(data=df_base[df_base['ano_final']==2020], x="mes_final", y="PREÇO MÉDIO REVENDA")
```

```python
plt.figure(figsize=(10, 6)) 
sns.lineplot(data=df_base[df_base['ano_final']==2020], x="mes_final", y="PREÇO MÉDIO REVENDA", hue='ESTADO')
```

```python
plt.figure(figsize=(10, 6)) 
sns.lineplot(data=df_base, x="ano_final", y="PREÇO MÉDIO REVENDA", hue='mes_final')
```

```python
table = pd.pivot_table(df_base, values='PREÇO MÉDIO REVENDA', index=['ano_final'], columns = ['PRODUTO'],
                       aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df_base, values='PREÇO MÉDIO REVENDA', index=['ESTADO'], columns = ['PRODUTO'],
                       aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df_base, values='PREÇO MÉDIO REVENDA', index=['REGIÃO'], columns = ['PRODUTO'],
                       aggfunc=np.sum)
table
```

```python
sns.countplot(data=df_base, x='REGIÃO', hue = 'PRODUTO')
```

```python
df_base_gasolina = df_base.loc[df_base['PRODUTO']=='GASOLINA COMUM']
df_base_gasolina.head()
```

```python
table = pd.pivot_table(df_base_gasolina, values='PREÇO MÉDIO REVENDA', index=['ano_final'], columns = ['REGIÃO'],
                       aggfunc=np.sum)
table
```

```python
plt.figure(figsize=(10, 6)) 
sns.countplot(data=df_base_gasolina, x='ano_final', hue = 'REGIÃO')
plt.xticks(rotation=45)
plt.show()
```
